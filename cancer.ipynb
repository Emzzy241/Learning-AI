```python
# importing our dataset

import pandas as pd

dataset = pd.read_csv('cancer.csv')
```

```python
# setting up our x and y attribute to predict whether or not the tumor is malignant or benine
# the x attribute will be all of our excel features. Now we will have our ai map out all of the correlations between this two features
# and that will allow the ai to predict whether or not a tumor is malignant or benine

# all we are saying here is our x attribute = everything in the dataset except for the column we wrapped inside it.
x = dataset.drop(columns=["diagnosis(1=m, 0=b)"])
```

```python
y = dataset["diagnosis(1=m, 0=b)"]
```

```python
# Now we have both our x and y column
# all that is left now is to split up our data between a training set and a testing set
# sklearn is a popular machine learning library
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
# so basically 20% of our data will be in the training set where we will be training our AI later

# NOw lets build our AI with Tensorflows Keras... We will also be building a neural network which is an adaptive ai and its used in any problem
```

```python
# lets first import tensor flow 
import tensorflow as tf
model = tf.keras.models.Sequential()
```

```python
# Next we need to start adding layers to our models
# the sigmoid function for our activation function was taken from the neural network,,, a sigmoid graph. 
model.add(tf.keras.layers.Dense(256, input_shape=x_train.shape[1:], activation='sigmoid'))
model.add(tf.keras.layers.Dense(256, activation='sigmoid'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```python
# Next is to compile our model
# we need an optimizer adam compiling(named adam)
# using binary_crossentropy because we are doing binary classification... This is very good when we're talking about categorical stuff just like our cancer's data
# the metric to look at is accuracy betcause we want to correctly classify as many tumors as possible

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

```

```python
model.fit(x_train, y_train, epochs=1000)
```

```python
# Lastly, we fit our models, we fit the x_train of our model to the y_train
# We also set the nu of epochs to 1000. Epochs mean the amount of times our algorithm is iterating over the same data

# Final thing to do in evaluating our algorithm to know the % of cancer detection our AI has.
# We will be evaluating with the x_test(our testing set) on the y_test

model.evaluate (x_test, y_test)

# so basically what the above is doing is its comparing what the moddel thinks that y_test
```

```python
# our accuracy is 97%, this just means that on a new data; our machine learning algotithm(our AI) can correctly classify cancer diagnoses, cancer tumors witha 97% accuracy. This is amazing; our Algorithm never went to College and it didn't go to school but it can detect cancer
# Our AI does not even know what a tumor is but from this data it was able to fnd a pattern and from that it can be used to help Radiologist and Doctors.


model.evaluate (x_test, y_test)
```
